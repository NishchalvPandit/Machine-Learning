```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
file_path = '/Users/nishchalpandit/Downloads/breast cancer prediction data.csv'
df = pd.read_csv(file_path)
print(df.head())
df.shape
```

```python
data=df.drop(['Unnamed: 32','id'], axis=1)
print(data.head())
data.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,1:], data.iloc[:,0],test_size=0.2, random_state=2)

print(X_train.head())
X_train.shape
```

```python
X_test.shape
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print(X_train)
```

```python
X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score

y_pred = knn.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
scores = []

for i in range(1,25):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    
    knn.fit(X_train,y_train)
    
    y_pred = knn.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))
print(scores)
    
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,25),scores)
```
